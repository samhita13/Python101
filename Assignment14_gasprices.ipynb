```python
#1. Read gasprices.txt file 
#2. Print each line of the gasprices.txt file in the 
#following format. Please ensure that you provide Exception 
#Handling to catch the error "File Not Found Exception'
try:
  #opem output file
  writeobj = open("output.txt" , "wt") 
  #open input file
  with open("gasprices.txt","rt") as fileobj :
    for eachline in fileobj:
      #split the date with -
      linelist=eachline.split("-")
      print("month:", linelist[0])
      writeobj.write("month:" + linelist[0]+ "\n")
      print("date:" , linelist[1])
      writeobj.write("date:"+ linelist[1] +"\n" )
      #split the price from year
      linelist2=linelist[2].split(":")
      print("year:", linelist2[0])
      writeobj.write("year:" + linelist2[0] + "\n" )
      print("price:" , linelist2[1])
      writeobj.write("price:" + linelist2[1] + "\n" )
  writeobj.close()
except FileNotFoundError:
  print('Error: please check if gasprices.txt is located in current directory')

```
