```python
# Create a text file with your own words (input.txt)
# Read the text file
# For each word you find, add it the dictionary (if it is not there in the dictionary)
#                         if it is there in the dictionary, simply update the COUNT
# at the end of the loop, print the KEYs (words) and the VALUE (count)

# Sample output
# Frequecy Table
# ant 4
# bat 3
# zebra 10
# snake 4
try:
  # Empty dictionary
  filedictionary = {}
  with open("goldilocks.txt", "rt") as fileobj:
    for eachline in fileobj:
      # list of words in a line
      linelist = eachline.split(" ")
      for word in linelist:
        # check the word in dictionary
        value = filedictionary.get(word)
        # word is not in dictionary.
        if value == None:
          value = 1
        else:
          # word is present, increase the count
          value = value + 1
        #update the new pair to dictionary
        filedictionary.update({word:value})
    for key in filedictionary:
      print(key,filedictionary[key] )
except FileNotFoundError:
  print('Error: please check if golidlocks.txt is located in current directory')
```
